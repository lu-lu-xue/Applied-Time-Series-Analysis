```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from numpy.fft import fft, ifft
```

Figure 4.1 
Time series plots of $(X^{(1)}_t), (X^{(2)}_t), (X_t) \text{ and } ( \widetilde X_t)$.

```python
t = np.arange(1, 49) 
x1 = np.cos(2*np.pi*t/12) + np.sin(2*np.pi*t/12)
x2 = 4*np.cos(2*np.pi*t/6) + 4*np.sin(2*np.pi*t/6)
x = x1 + x2
tildex = x + np.random.normal(size=48)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,7))
axs[0, 0].plot(t, x1)
axs[0, 0].set_title('Xt1')
axs[0, 1].plot(t, x2, 'tab:orange')
axs[0, 1].set_title('Xt2')
axs[1, 0].plot(t, x, 'tab:green')
axs[1, 0].set_title('Xt')
axs[1, 1].plot(t, tildex, 'tab:red')
axs[1, 1].set_title('tilde Xt')

fig.tight_layout()

#for ax in axs.flat:
#    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
#for ax in axs.flat:
#    ax.label_outer()
```

Figure 4.2 

The scaled periodograms of $(X_t), (\widetilde X^{(1)}_t) \text{ and } (\widetilde X^{(2)}_t)$.

```python
tildex2 = x + np.random.normal(scale=3, size=48)
```

```python
# calculate (scaled) periodogram
def periodogram(x):
    I = np.abs(fft(x)/np.sqrt(48))**2
    P = 4*I/48
    f = np.arange(0,25)/48

    return (P,f)
```

```python
P_x, f_x = periodogram(x)
P_tildex1, f_tildex1 = periodogram(tildex)
P_tildex2, f_tildex2 = periodogram(tildex2)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10,4))
axs[0].plot(f_x, P_x[0:25], '-')
axs[0].set_title('Xt')
axs[1].plot(f_tildex1, P_tildex1[0:25], '-')
axs[1].set_title('tilde Xt1')
axs[2].plot(f_tildex2,P_tildex2[0:25], '-')
axs[2].set_title('tilde Xt2')
```
