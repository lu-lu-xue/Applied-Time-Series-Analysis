```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Figure 1.5

```python
# plot for cyclical time series
def genCTS(R,phi):
    t = np.arange(1, 101)  
    # R = np.random.uniform(-0.5, 1, 100)  
    # phi = np.random.uniform(0, 1, 100)  
    lmbda = np.pi / 25  

    cyc = R * np.sin(lmbda * t + phi)  

    plt.plot(cyc)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title('Cyclical Time Series')
    plt.show()
```

```python
R = np.random.uniform(-0.5, 1, 100)  
phi = np.random.uniform(0, 1, 100)
genCTS(R,phi)
```

```python

R = np.random.uniform(-0.3, 1, 100)
phi = np.random.uniform(0, 1, 100)
genCTS(R,phi)
```

```python

R = np.random.uniform(-0.5, 1, 100)  
phi = np.random.standard_normal(100)
genCTS(R,phi)
```

Figure 1.6

```python
z = np.random.standard_normal(100)
plt.plot(z)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('I.I.D. Noise')
plt.show()
```

```python
rw = np.cumsum(z)
plt.plot(rw)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Random Walk')
plt.show()
```
