Figure 1.5

100 simulated values of the cyclical time series (left panel), the stochastic amplitude (middle
panel), and the sine part (right panel).

```r
# cyclical ts
t = 1:100; R = runif(100,-.5,1); phi = runif(100,0,1); lambda = pi/25
cyc = R*sin(lambda*t+phi)

# stochastic amplitude
R2 = runif(100, -1, 1)
cyc2 = R2*sin(lambda*t+phi)

# sine part
phi3 = rnorm(100)
cyc3 = R*sin(lambda*t+phi3)


par(mfrow=c(1,3))
plot.ts(cyc, xlab="", ylab="")
plot.ts(cyc2, xlab="", ylab="")
plot.ts(cyc3, xlab="", ylab="")
```

Figure 1.6

1000 simulated values of iid $N(0; 1)$ noise (left panel) and a random walk with iid $N(0; 1)$
innovations (right panel).

```r
par(mfrow=c(1,2))

z = rnorm(1000, 0, 1)

rw = cumsum(z)

plot.ts(z, xlab="", ylab="", main="")
plot.ts(rw, xlab="", ylab="")
```
