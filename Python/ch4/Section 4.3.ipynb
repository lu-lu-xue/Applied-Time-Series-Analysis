```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
from numpy.fft import fft, ifft
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
# use airline passengers dataset
filename = "../../data/" + "airpassengers.csv"
data = pd.read_csv(filename)
passengers = data['Passengers'].values
```

```python
data.head()
```

```python
plt.plot(passengers)
plt.ylabel("Passenger numbers ('000)")
plt.title("Air Passengers from 1949 to 1961")
```

```python
f, Pxx_den = signal.periodogram(passengers,fs=1/48,scaling='density')
plt.plot(f,Pxx_den)
plt.xlabel('frequency [Hz]')
#plt.ylabel('PSD [V**2/Hz]')
plt.show()
```

```python
f, Pxx_den = signal.periodogram(passengers,fs=1/4,scaling='density')
plt.plot(f,Pxx_den)
plt.xlabel('frequency [Hz]')
plt.show()
```
