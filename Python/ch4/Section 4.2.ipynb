```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft
from statsmodels.tsa.arima_process import ArmaProcess
```

Figure 4.3 

Time series plots of white noise $(Z_t:t \in \mathbb{Z})$ (left), two-point moving average $(X_t:t\in \mathbb{Z})$ (middle) and spectral density of $(Z_t:t\in \mathbb{Z})$ (right).

```python
# white noise
zt = np.random.uniform(0, 1, 100)

zt_1 = zt[1:]
zt_1 = np.append(zt_1, 0)
xt = 1/2 * (zt + zt_1)

w = np.arange(0,0.5, 0.0001)
fxw = 1/2 * (1 + np.cos(2 * np.pi * w))
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12,4))
axs[0].plot(zt)
axs[0].set(ylabel='z')
axs[1].plot(xt)
axs[1].set(ylabel='y')
axs[2].plot(w, fxw)
```

Figure 4.4

Time series plot and spectral density of the AR(2) process in Example 4.2.3.

```python
sigma2 = 89.34
phi1 = 1.35
phi2 = -0.41

ar = np.array([1, -phi1, -phi2])
ma = np.array([1])
simulatedata = ArmaProcess(ar, ma).generate_sample(nsample=200)

fxw_44 = sigma2 * np.reciprocal(((1+phi1**2+phi2**2)+2*(phi1 * phi2 - phi1)*np.cos(2*np.pi*w)-2*phi2*np.cos(4*np.pi*w)))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,4))
axs[0].plot(simulatedata)
#axs[0].set(ylabel='z')
axs[1].plot(w,fxw_44)
#axs[1].set(ylabel='y')
```
