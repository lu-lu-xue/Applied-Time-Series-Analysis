```python
import numpy as np
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
```

```python
# create ARMA
def createARMA(ar_l,ma_l):
    ar = np.array(ar_l)
    ma = np.array(ma_l)

    return (ar,ma)
```

```python
# define ARMA process
def geAarmaSamples(ar,ma,n_samples):
    return ArmaProcess(ar, ma).generate_sample(nsample=n_samples)
```

```python
def plotARMA(simulated_data,p,q):
    plt.plot(simulated_data)
    plt.title(f"Simulated ARMA({p},{q}) Process")
    plt.xlim([0, 100])
    plt.show()
```

Figure 3.1

Realizations of three autoregressive moving average processes.

```python
n_samples = 100
```

```python
# Define the ARMA parameters
ar_arr = [1, 0.2, -0.3]
ma_arr = [1, -0.5, 0.3]

ar,ma = createARMA(ar_arr,ma_arr)
simulated_data = geAarmaSamples(ar,ma,n_samples)

plotARMA(simulated_data,2,2)
```

```python
ar_arr = [1, 0.3]
ma_arr = [1, -0.2, -0.3, 0.5, 0.2]

ar,ma = createARMA(ar_arr,ma_arr)
simulated_data = geAarmaSamples(ar,ma,n_samples)

plotARMA(simulated_data,1,4)
```

```python
ar_arr = [1, -0.2, -0.3, 0.5, 0.2]
ma_arr = [1, 0.6]

ar,ma = createARMA(ar_arr,ma_arr)
simulated_data = geAarmaSamples(ar,ma,n_samples)

plotARMA(simulated_data,4,1)
```
